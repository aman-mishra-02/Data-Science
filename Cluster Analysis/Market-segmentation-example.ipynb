# Market segmentation example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv ('3.12.Example.csv')
data.head()
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
kmeans.predict(x)
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = kmeans.predict(x)
```

```python
plt.scatter(data_with_clusters['Satisfaction'], data_with_clusters['Loyalty'], c=data_with_clusters['Clusters'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
# Standardizing 

from sklearn import preprocessing

scaled = preprocessing.scale(x)
```

```python
x.shape
```

```python
wcss=[]
for i in range(1,30) :
    kmeans = KMeans(i)
    kmeans.fit(scaled)
    wcss.append(kmeans.inertia_)

```

```python
i = range(1,30)

plt.plot(i,wcss)
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(scaled)
```

```python
data_with_clusters2 = data.copy()
data_with_clusters2['Clusters'] = kmeans_new.predict(scaled)
```

```python
plt.scatter(data_with_clusters2['Satisfaction'], data_with_clusters2['Loyalty'], c=data_with_clusters2['Clusters'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
