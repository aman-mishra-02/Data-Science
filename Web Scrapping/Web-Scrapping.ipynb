```python
from bs4 import BeautifulSoup
import requests 
```

```python
HTML_Text = requests.get('https://letterboxd.com/aman127/films/diary/').text
print(HTML_Text)
```

```python
soup = BeautifulSoup(HTML_Text, 'lxml')
```

```python
Movie = soup.find_all('h3', class_='headline-3 prettify')
for m in Movie:
    print(m.a.text)
```
