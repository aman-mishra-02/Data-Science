```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(-10,10,size=(1000,1))
Z = np.random.uniform(-10,10,size=(1000,1))

inputs = np.column_stack((X,Z))

inputs.shape
```

```python
targets = 2*X + 3*Z + 5 + np.random.uniform(-1,1,size=(1000,1))

targets.shape
```

```python
init_range = 0.1

weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))


biases = np.random.uniform(low=-init_range, high=init_range, size=1)

#Print the weights to get a sense of how they were initialized.
print (weights)
print (biases)
```

```python
learning_Rate = 0.03
```

```python
for i in range(1000):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    loss = np.sum(deltas ** 2) / 2 / 1000
    print (loss)
    deltas_scaled = deltas / 1000

    # Gradient Function
    weights = weights - learning_Rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_Rate * np.sum(deltas_scaled)
    
```

```python
plt.scatter(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
